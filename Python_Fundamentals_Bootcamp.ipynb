**Question:**

You have been asked, as a junior software developer, to write a Python program that processes the performance of one trainee and generates a simple performance summary for the training manager.

1.	Capture the trainee's full name.

4.	Calculate the average score.
5.	Assign a performance grade using the criteria below:
o	80 and above = Excellent

o	50 to 59 = Fair

o	If the average is 50 or above, the trainee is marked as Competent
o	Otherwise, the trainee is marked as Not Yet Competent

o	Trainee name
o	Scores in all assessment areas
o	Total score
o	Average score
o	Performance grade
o	Completion status
